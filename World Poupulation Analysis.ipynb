# Visualization of Population

```python
#Least populated country in 2022

plt.subplots(figsize=(20,8))
plt.tight_layout(pad=7.0)

p=sns.barplot(x=df["Country/Territory"],y=df["2022 Population"],order=df.sort_values("2022 Population",ascending=True)["Country/Territory"][:11],saturation=1,edgecolor = "#1c1c1c", linewidth = 4)
p.set_yscale("linear")
p.set_title("Least Populated Countries In 2022",fontsize=25)
p.set_xlabel("Country",fontsize=20)
p.set_ylabel("Population",fontsize=20)
p.set_xticklabels(p.get_xticklabels(),rotation = 12)
for container in p.containers:
    p.bar_label(container,label_type="edge",padding=6,size=18,color="black",rotation=0)


sns.despine(left=True, bottom=True)
plt.show()


```

```python
#Most populated country in 2022

plt.subplots(figsize=(20,8))
plt.tight_layout(pad=7.0)
p=sns.barplot(x=df["Country/Territory"],y=df["2022 Population"],order=df.sort_values("2022 Population",ascending=False)["Country/Territory"][:11], saturation=1,edgecolor = "#1c1c1c", linewidth = 4)
p.set_yscale("log")
p.set_title("Top Populated Countries In 2022",fontsize=25)
p.set_xlabel("Country",fontsize=20)
p.set_ylabel("Population",fontsize=20)
p.set_xticklabels(p.get_xticklabels(),rotation = 0)
for container in p.containers:
    p.bar_label(container,label_type="edge",padding=6,size=18,color="black",rotation=0)
                     

sns.despine(left=True, bottom=True)
plt.show()
```

```python
#Continent population in 2020
df1 = df.copy()
df1 = pd.DataFrame(df1.groupby(["Continent"])["1970 Population","1980 Population","1990 Population","2000 Population","2010 Population","2015 Population","2020 Population","2022 Population"].sum())

print("Let's have a look on the continent-wise population in 2022 :")
plt.subplots(figsize=(20,8))
p=sns.barplot(x=df1.index, y=df1["2022 Population"],order=df1.sort_values("2022 Population",ascending=False).index, saturation=1,edgecolor = "#1c1c1c", linewidth = 4)
p.set_yscale("log")
p.set_title("Continent Population in  2022",fontsize=25)
p.set_xlabel("\nContinent",fontsize=20)
p.set_ylabel("Population",fontsize=20)
p.set_xticklabels(p.get_xticklabels(),rotation = 0)
for container in p.containers:
    p.bar_label(container,label_type="edge",padding=6,size=18,color="black",rotation=0)

sns.despine(left=True, bottom=True)
plt.show()
```

```python
plt.subplots(figsize=(12, 12))

labels = "Asia","Africa","Europe","North America","South America","Oceania"
size = 0.5

wedges, texts, autotexts = plt.pie([df1.sort_values("2022 Population",ascending=False)["2022 Population"][0],
                                    df1.sort_values("2022 Population",ascending=False)["2022 Population"][1],
                                    df1.sort_values("2022 Population",ascending=False)["2022 Population"][2],
                                    df1.sort_values("2022 Population",ascending=False)["2022 Population"][3],
                                    df1.sort_values("2022 Population",ascending=False)["2022 Population"][4],
                                    df1.sort_values("2022 Population",ascending=False)["2022 Population"][5]],
                                    explode = (0,0,0,0,0,0),
                                    textprops=dict(size= 20, color= "white"),
                                    autopct="%.2f%%", 
                                    pctdistance = 0.72,
                                    radius=.9, 
                                    
                                    wedgeprops=dict(width = size, edgecolor = "black", 
                                    linewidth = 4),
                                    startangle = -15)

plt.legend(wedges, labels, title="Continent",loc="center left",bbox_to_anchor=(1, 0, 0, 1), edgecolor = "black")
plt.title("\nContinent Population Ratio 2022",fontsize=25)
plt.show()
```

# Continent Population Density


```python
df_den = df.copy()
df_den = pd.DataFrame(df_den.groupby(["Continent"])["1970 Population","1980 Population","1990 Population","2000 Population","2010 Population","2015 Population","2020 Population","2022 Population","Area (km²)"].sum())
col = ["1970 Population","1980 Population","1990 Population","2000 Population","2010 Population","2015 Population","2020 Population","2022 Population"]
for i in col:
    df_den[i] = df_den[i]/df_den["Area (km²)"]    
df_den.rename(columns={"1970 Population":"1970 Density","1980 Population":"1980 Density","1990 Population":"1990 Density","2000 Population":"2000 Density","2010 Population":"2010 Density","2015 Population":"2015 Density","2020 Population":"2020 Density","2022 Population":"2022 Density"},inplace=True)
df_den.drop(columns="Area (km²)",inplace=True)


print(f"Let's have a look on the timeline of continent-wise population density :")
_, axs = plt.subplots(figsize=(20,10))

sns.lineplot(x=df_den.T.index ,y=df_den.T["Asia"],data=df_den.T,ax=axs,color="#006837",marker="o",linewidth=5,markersize=20)
sns.lineplot(x=df_den.T.index ,y=df_den.T["Africa"],data=df_den.T,ax=axs,color="#66BD63",marker="o",linewidth=5,markersize=20)
sns.lineplot(x=df_den.T.index ,y=df_den.T["Europe"],data=df_den.T,ax=axs,color="#D9EF8B",marker="o",linewidth=5,markersize=20)
sns.lineplot(x=df_den.T.index ,y=df_den.T["North America"],data=df_den.T,ax=axs,color="#FEE08B",marker="o",linewidth=5,markersize=20)
sns.lineplot(x=df_den.T.index ,y=df_den.T["South America"],data=df_den.T,ax=axs,color="#F46D43",marker="o",linewidth=5,markersize=20)
sns.lineplot(x=df_den.T.index ,y=df_den.T["Oceania"],data=df_den.T,ax=axs,color="#A50026",marker="o",linewidth=5,markersize=20)

axs.set_title("Continent-wise Population Density Timeline\n",fontsize=25)
axs.set_xlabel("\nYear",fontsize=20)
axs.set_ylabel("Population Density",fontsize=20)
axs.legend(["Asia","Africa","Europe","North America","South America","Oceania"],title="Continent", edgecolor = "#1c1c1c")
# axs.set_xticks([],minor=False)

sns.despine(left=True, bottom=True)
plt.show()
```
